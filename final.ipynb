```python
# In the Exploration Data process, I indentified outliers in the 'hum' and 'windspeed' column. Now I'm going to remove those outliers
```

```python
# First for 'hum'

# Calculate the quartiles and IQR for 'hum' column
Q1_hum = bike_data['hum'].quantile(0.25)
Q3_hum = bike_data['hum'].quantile(0.75)
IQR_hum = Q3_hum - Q1_hum

# Define lower and upper bounds
lower_bound_hum = Q1_hum - 1.5 * IQR_hum
upper_bound_hum = Q3_hum + 1.5 * IQR_hum

# Remove outliers in the 'hum' column
bike_data = bike_data[(bike_data['hum'] >= lower_bound_hum) & (bike_data['hum'] <= upper_bound_hum)]
```

```python
# And then for 'windspeed'

# Calculate the quartiles and IQR for 'windspeed' column
Q1_windspeed = bike_data['windspeed'].quantile(0.25)
Q3_windspeed = bike_data['windspeed'].quantile(0.75)
IQR_windspeed = Q3_windspeed - Q1_windspeed

# Define lower and upper bounds
lower_bound_windspeed = Q1_windspeed - 1.5 * IQR_windspeed
upper_bound_windspeed = Q3_windspeed + 1.5 * IQR_windspeed

# Remove outliers in the 'windspeed' column
bike_data = bike_data[(bike_data['windspeed'] >= lower_bound_windspeed) & (bike_data['windspeed'] <= upper_bound_windspeed)]
```

```python
# This shows that there are no outliers anymore in those columns
sns.boxplot(data=bike_data[['hum', 'windspeed']])
plt.title('Box Plots of Humidity, and Windspeed')
plt.show()
```

```python
# Perform one-hot encoding
bike_data = pd.get_dummies(bike_data, columns=['weekday'])
```

```python
from sklearn.model_selection import train_test_split

X = bike_data.drop(columns=['cnt'])
y = bike_data['cnt']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Combining the target variable and all relevent features
final_dataset = pd.concat([X, y], axis=1)
```

```python
final_dataset
```

```python
# Decision tree

from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Initialize and train the Decision Tree model
decision_tree_model = DecisionTreeRegressor(random_state=42)
decision_tree_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = decision_tree_model.predict(X_test)

# Evaluate the model's performance using Mean Squared Error and R-squared
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Decision Tree MSE:", mse)
print("Decision Tree R-squared:", r2)
```

```python
# So in this Decision tree, we calculated the mean squared error and r-squared to check the performance.
# Because MSE is big, it means that there are large prediction errors, and that the predictions are not a perfect fit to actual data
# But R2 is relativly good as it means that the model explains 78.75% of varience in the bike rental counts and there is some unexplained varience
```
